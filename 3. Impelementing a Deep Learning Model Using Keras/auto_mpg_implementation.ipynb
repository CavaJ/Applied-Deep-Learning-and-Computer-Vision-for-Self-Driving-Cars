```python
#Importing important libraries

import numpy as np
import pathlib
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras import layers
#from __future__ import absolute_import, division, print_function 
print(tf.__version__)
```

```python
# Datapath to import auto-mpg data
dataset_path = keras.utils.get_file("auto-mpg.data", "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
```

```python
# Coloumn names for dataset
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin'] # omitted the car name column
```

```python
#Importing dataset
raw_dataset = pd.read_csv(dataset_path, names=column_names,
                      na_values = "?", comment='\t',
                      sep=" ", skipinitialspace=True)
```

```python
# Checking the dataset

raw_dataset.head()
```

```python
#Creating a copy of dataset
dataset = raw_dataset.copy()
```

```python
# Is there any NA value in data frame? 

dataset.isna().sum()
```

** We have observed there is 6 NA value in the the Horsepower column 

```python
# NA values are less so we can drop it
dataset = dataset.dropna()
```

```python
# check NA values again
dataset.isna().sum()
```

** We can observe that column "Origin" is a categorical variable so we create one hot encoding

```python
origin = dataset.pop('Origin')
dataset['USA'] = (origin == 1)*1.0
dataset['Europe'] = (origin == 2)*1.0
dataset['Japan'] = (origin == 3)*1.0
dataset.tail() # Return the last `n` rows - default 5.
```

** There are 3 countries, so we can observe encoding for 3 countries are created

```python
#Creating train and test samples of 80/20 split
train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
#Multivariate analysis 
sns.pairplot(train_dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind="kde")
```

```python
#Creating train and test labels
train_labels = train_dataset.pop('MPG') # returns values from the selected column and drops that column
test_labels = test_dataset.pop('MPG')
```

```python
# usually, when we use neural networks, we get improved performance when we standardize the data. 
# standardization just means normalizing the values so that they all fit between a certain range, such as 0 to 1 or -1 to +1. scikitlearn.MinMaxScaler(feature_range=(0, 1)) and scikitlearn.MinMaxScaler(feature_range=(-1, 1)).

#Method to normalize (standardize) the data, with mean=0 and std=1, turns the data into normal distribution.
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
```

```python
#Normalizing train and test data
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

#### TIP: In general, it is recommended to normalize the data in one go and split it into train and test sets. In that way, the data in train and test sets will come from the same data distribution.

# Building and compiling the model

```python
len(train_dataset.keys()) # number of columns (Cylinders, Displacement, Horsepower, Weight, Acceleration, Model Year, USA, Europe, Japan)
```

```python
# Method to build model 
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation=tf.nn.relu, input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation=tf.nn.relu),
    layers.Dense(1) # output layer for mpg prediction
  ])
  optimizer = tf.keras.optimizers.Adam(0.001)
 
  model.compile(loss='mean_squared_error', # mean squared error is a common loss function used for regression problems
                optimizer=optimizer,
                metrics=['mean_absolute_error', 'mean_squared_error']) # mean absolute error and mean squared error are the evaluation metrics for regression problems
  return model
```

```python
#Creating instance of a model
model = build_model()
```

```python
# Model Summary
model.summary() # 64 * 9 + 64 (a bias for every neuron) = 640 params - first layer.
```

```python
#Creating examples
example_batch = normed_train_data[:10]
```

```python
example_batch.shape
```

```python
#Predicting on example batch
example_result = model.predict(example_batch)
example_result
```

```python
# the patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10) # patience: number of epochs with no improvement after which training will be stopped.
```

```python
# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')
```

```python
#Number of epochs
EPOCHS = 2000
```

```python
# If your test set is meant for final model evaluation, you should still use a validation set during training. In this case, setting validation_split=0.2 is a good idea because:
# - The model can tune hyperparameters (e.g., learning rate, architecture, early stopping) using the validation set.
# - The test set remains untouched until final evaluation, ensuring an unbiased assessment.

# If you manually split your dataset into train, validation, and test sets, then do NOT use validation_split. Instead, pass your validation set explicitly using validation_data:
# model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))
# This ensures your predefined X_val and y_val are used correctly as the validation set.
```

```python
#Training model
history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0, # validation_split: Float between 0 and 1. Fraction of the training data to be used as validation data. The model will set apart this fraction of the training data, will not train on it, and will evaluate the loss and any model metrics on this data at the end of each epoch. The validation data is selected from the last samples in the `x` and `y` data provided, before shuffling.
  # callbacks=[PrintDot()]
  callbacks=[early_stop, PrintDot()]
)
```

```python
#Lets check loss, mean absolute error, mean square error, val loss, val mean squared error and val mean absolute error
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail(10)
```

```python
# Method to Plot loss

def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch
  
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mean_absolute_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_absolute_error'],
           label = 'Val Error')
  plt.ylim([0,5])
  plt.legend()
  
  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mean_squared_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_squared_error'],
           label = 'Val Error')
  plt.ylim([0,20])
  plt.legend()
  plt.show()
```

```python
#Plot Train and validation loss
plot_history(history)
```

# Evaluate the model

```python
test_loss, test_mae, test_mse = model.evaluate(normed_test_data, test_labels, verbose=0)
print("Testing dataset Mean Abs Error (MAE): {:5.2f} MPG".format(test_mae))
```

# Make predictions

```python
# Performing prediction 

test_predictions = model.predict(normed_test_data).flatten()

# Plot true values in red and predictions in orange
plt.scatter(test_labels, test_labels, color='red', label="True Values")  # True values in red
plt.scatter(test_labels, test_predictions, color='orange', label="Predictions")  # Predictions in orange

plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0, plt.xlim()[1]])
plt.ylim([0, plt.ylim()[1]])

# Add a perfect prediction line (diagonal)
plt.plot([-100, 100], [-100, 100], 'k--', label="Perfect Fit") 

plt.legend()  # Show the legend
plt.show()
```

```python
# Checking the loss
error = test_predictions - test_labels
# plot the error distribution; we can expect non-gaussian error distribution since the number of samples in auto-mpg dataset is small-around 400.
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]")
_ = plt.ylabel("Count")
```
